```python
def tree(var, a):
    tree_depth = len(a)
    for index in range(tree_depth):
        if a[index] == "female":
            var = var.loc[var['Sex'] == "female"]
        if a[index] == "male":
            var = var.loc[var['Sex'] == "male"]
        if a[index] == "surv":
            var = var.loc[var['Survived'] == 1]
        if a[index] == "norsurv":
            var = var.loc[var['Survived'] == 0]
        if a[index] == "adult":
            var = var.loc[var['Age'] > 15]  
        if a[index] == "child":
            var = var.loc[var['Age'] < 16] 
        if a[index] == "young":
            var = var.loc[var['Age'] < 50] 
        if a[index] == "old":
            var = var.loc[var['Age'] > 49]
        if a[index] == "fclass":
            var = var.loc[var['Pclass'] == 1] 
        if a[index] == "mclass":
            var = var.loc[var['Pclass'] == 2] 
        if a[index] == "lclass":
            var = var.loc[var['Pclass'] == 3] 
        if a[index] == "bro_sis":
            var = var.loc[var['SibSp'] != 0] 
        if a[index] == "Pa_ch":
            var = var.loc[var['Parch'] != 0] 
        if a[index] == "noPa_ch":
            var = var.loc[var['Parch'] == 0] 
        if a[index] == "zerofare":
            var = var.loc[var['Fare'] == 0] 
        if a[index] == "nozerofare":
            var = var.loc[var['Fare'] != 0] 
    return var
```

```python
def pclass_fare_describe(class_layer):
    plcass_nozerofare = tree(full, [class_layer, "nozerofare"])
    means = plcass_nozerofare["Fare"].mean(skipna=True)
    mins = plcass_nozerofare["Fare"].min(skipna=True)
    maxs = plcass_nozerofare["Fare"].max(skipna=True)
    return means, mins, maxs
```
